# 자연어 처리 개요: 코퍼스에서 언어 이해까지

## 자연어 처리란

자연어처리(NLP, Natural Language Processing)란 컴퓨터가 인간의 언어를 이해하고 생성하며 활용할 수 있도록 하는 기술이다.

텍스트나 음성 데이터를 분석하고 처리하는 데 사용되며, 예로는 번역, 감정 분석, 질의응답 시스템, 음성 인식 등이 있다.

    * 이미지/비디오 설명 Image/Video Captioning




## 02. 코퍼스와 데이터셋 준비

**코퍼스(corpus)** 는 특정 목적을 위해 모은 실제 언어 자료의 집합이다. 한 행이 문장인지 문서인지, 작성 시점·작성자·언어·레이블이 무엇인지를 먼저 정의해야 분석 단위가 흔들리지 않는다.

### 확인할 데이터 속성

- **단위(granularity):** 문서, 문단, 문장, 발화, 토큰 중 한 행이 무엇인지 확인한다.
- **형식과 인코딩:** 문자열 dtype, UTF-8 여부, 줄바꿈·구분자·따옴표 규칙을 확인한다.
- **품질:** 결측치, 중복, 깨진 문자, HTML, 개인정보, 클래스 불균형을 확인한다.
- **출처와 권리:** 수집 목적, 이용 약관, 라이선스, 개인정보 처리 범위를 기록한다.

### 자주 쓰는 파일 형식

- `txt`: 원문 또는 한 줄 한 문서. 단순하지만 레이블·메타데이터 표현이 약하다.
- `csv/tsv`: 표 형태 문서와 레이블. 본문 안 쉼표·탭·줄바꿈의 quoting을 확인한다.
- `json/jsonl`: 중첩 메타데이터 또는 한 줄 한 레코드. 대규모 스트리밍에는 JSONL이 편리하다.
- `xml/html`: 태그 구조를 보존하지만 본문 추출에는 XML/HTML 파서가 필요하다.
- `parquet`: dtype을 보존하고 압축·열 단위 읽기가 가능해 큰 데이터셋에 적합하다.

### 공개 코퍼스 자원 예시

- 국립국어원 모두의 말뭉치: 한국어 문어·구어·형태·구문 등(신청 조건과 라이선스 확인)
- AI Hub: 한국어 대화·감성·질의응답 등(데이터별 이용 조건 확인)
- NSMC: 한국어 영화 리뷰 감성 분류
- NLTK Corpus / Hugging Face Datasets: 영어 및 다국어 교육·연구 데이터

공개 데이터도 자동으로 자유 배포 가능한 것은 아니다. 수업 자료에 원문을 포함하기 전에 현재 라이선스와 개인정보 조건을 반드시 확인한다.

### 데이터셋 준비 절차

1. 과제와 예측 단위, 입력과 정답을 정의한다.
2. 출처·라이선스·개인정보 처리 근거를 확인한다.
3. 원본은 읽기 전용으로 보존하고 작업 사본을 만든다.
4. 인코딩·스키마·결측·중복·길이·레이블 분포를 프로파일링한다.
5. 정제·정규화 규칙을 학습/평가 데이터에 일관되게 적용한다.
6. 데이터 누수를 막도록 Train/Validation/Test를 분리하고 버전을 기록한다.


## 03. 주요 용어

### 말뭉치(Corpus)

- 자연어 처리에서 분석하거나 학습에 사용하는 텍스트 데이터의 모음이다.
- 특정 주제의 뉴스 기사, 소셜 미디어 게시글, 법률 문서 등이 있다.
- 일반적으로 특정 도메인에 따라 구성하며, 크고 다양한 데이터를 포함하는 것이 이상적이다.
- 과거에는 품사와 형태소 같은 보조 정보를 추가한 구조화 자료도 말뭉치에 포함하기도 했다.

### 토큰(Token)

- 자연어 문서를 분석하기 위해 긴 문자열을 작은 단위로 나눈 결과이다.
- 단어, 문자, 구두점 또는 서브워드가 토큰이 될 수 있다.
- 문장 `I love NLP!`는 `["I", "love", "NLP", "!"]`처럼 나눌 수 있다.
- 토큰화(Tokenization)는 문자열을 여러 토큰으로 분리하는 과정이다.
- 대표적인 특수 토큰은 `<START>`, `<EOS>`, `<UNK>`, `<PAD>`이다.

### 문장(Sentence)

- 문장부호와 문법 구조를 기준으로 구분하는 문맥 단위이다.
- `NLP is fascinating. I enjoy learning it.`은 두 문장으로 나뉜다.
- 문장 분리(Sentence Segmentation)는 구두점을 기본 기준으로 사용하지만 언어별 예외를 처리해야 한다.

### 단어(Word)

- 문장에서 독립적인 의미를 가지는 언어 단위이다.
- `I love NLP.`의 단어는 `["I", "love", "NLP"]`이다.
- 영어는 주로 공백과 구두점으로 구분하지만 언어에 따라 분리 규칙이 달라진다.
  - 중국어와 일본어는 띄어쓰기가 없어 별도 분리 처리가 필요하다.
  - 한국어는 조사와 어미가 결합하므로 형태소 분석이나 서브워드 분리가 중요하다.

### 형태소(Morpheme)

- 의미를 가지는 가장 작은 언어 단위이다.
- 영어의 `unbreakable`은 `un-`, `break`, `-able`로 나눌 수 있다.
  - `un-`: 부정 의미의 접두사
  - `break`: 어근
  - `-able`: 가능 의미의 접미사
- 형태소 분석은 어근과 접사를 구분하며 한국어처럼 형태 변화가 많은 언어에서 중요하다.

### 불용어(Stop Words)

- 자주 등장하지만 특정 분석 과제에 기여하는 정보가 적은 단어이다.
- 영어의 `I`, `my`, `me`, `over`나 한국어의 일부 조사·접미사가 예가 될 수 있다.
- NLTK는 영어 불용어 목록을 제공하지만, 실제 제거 여부는 과제와 도메인에 맞게 결정한다.


## 04. 언어 분석 층위와 자연어 이해(NLU)

NLP 시스템은 보통 **입력 수집 → 전처리 → 언어 표현 → 이해/예측 → 생성 또는 서비스 출력 → 평가**로 연결된다. NLU(Natural Language Understanding)는 이 가운데 텍스트의 의도, 개체, 관계, 의미를 구조화하는 영역이다. 반대로 NLG는 구조화된 정보나 문맥에서 자연어를 생성한다.

언어 분석 층위는 서로 연결된다.

1. **형태 분석:** 단어를 형태소와 품사로 나눈다. 예: `먹었다 → 먹/VV + 었/EP + 다/EF`
2. **구문 분석:** 단어 사이의 문법적 관계를 찾는다. 예: 누가 무엇을 먹었는가
3. **의미 분석:** 문장 표현을 개체·관계·의도·논리 의미로 해석한다.
4. **담화 통합:** 여러 문장 사이의 지시 대상과 연결 관계를 해석한다.
5. **화용 분석:** 발화 상황, 화자 의도, 상식과 사회적 맥락을 반영한다.

### CFG로 보는 문장 구조

CFG(Context-Free Grammar)는 `비단말 기호 → 기호열` 규칙으로 문장 구조를 표현한다.

```text
S  -> NP VP
NP -> N | DET N
VP -> V NP
```

`학생이 책을 읽는다`를 주어 명사구(NP)와 서술어구(VP)로 나누는 식이다. 규칙 기반 분석은 구조를 설명하기 쉽지만 실제 언어의 예외와 모호성을 모두 규칙으로 작성하기 어렵다. 현대 NLP는 통계·딥러닝 모델을 주로 사용하되, CFG는 구문 트리와 구조적 모호성을 설명하는 좋은 기준이다.


## 05. 모호성·담화·화용: 문장만 읽어서는 부족한 정보

- **어휘 모호성:** `배`가 과일, 신체, 선박 중 무엇인지 문맥으로 정한다.
- **구문 모호성:** `망원경으로 사람을 봤다`에서 망원경을 누가 가졌는지 구조가 둘 이상일 수 있다.
- **의미 모호성:** 수식 범위, 부정, 양화 표현에 따라 전체 뜻이 달라질 수 있다.
- **담화:** `민수가 우산을 샀다. 그는 바로 펼쳤다.`에서 `그`의 지시 대상을 앞 문장과 연결한다.
- **화용:** `창문이 열려 있네요`는 상황에 따라 사실 진술이 아니라 닫아 달라는 요청일 수 있다.

따라서 좋은 NLP 데이터는 문장 텍스트만이 아니라 필요한 경우 대화 순서, 화자, 시간, 문서 출처 같은 문맥 메타데이터도 보존해야 한다.

## 07. Text Vectorization: 표현 단위와 생성 방식

텍스트 벡터화(Text Vectorization)는 사람이 읽는 텍스트를 모델이 계산할 수 있는 숫자 표현으로 변환하는 과정이다. 벡터화 기법을 구분할 때는 다음 두 질문을 분리해야 한다.

1. **무엇을 하나의 벡터로 표현하는가?** 문서·문장·토큰 가운데 표현 단위를 정한다.
2. **벡터의 값을 어떻게 만드는가?** 빈도, 학습된 고정 벡터, 문맥을 반영한 벡터 가운데 생성 방식을 정한다.

[단어 표현을 지역 표현과 연속 표현으로 나눈 전체 지도]
- 희소한 지역 표현에서 학습된 밀집 표현으로 발전하는 큰 흐름을 확인

### 07-01. 모델에 따라 달라지는 벡터화 흐름

#### 전통적인 머신러닝

```text
문서 → 토큰화 → 어휘 사전 → BoW/TF-IDF 벡터 → 머신러닝 모델
```

> **BoW(Bag of Words):** 단어 순서를 무시하고 문서에 등장한 단어의 빈도를 벡터로 표현하는 방법이다.

> **TF-IDF(Term Frequency-Inverse Document Frequency):** 한 문서에는 자주 등장하지만 전체 문서에서는 드문 단어에 높은 가중치를 주는 방법이다.

- 문서 하나가 특성 행렬의 한 행이 된다.
- 일반적인 행렬의 형태는 `(문서 수, 어휘 수)`이다.
- 단어가 많아질수록 대부분의 값이 0인 희소 행렬이 만들어진다.

아래 그림은 각 문서에서 단어가 등장한 횟수를 세어 문서-단어 행렬로 만드는 BoW의 결과를 보여 준다. 행은 문서, 열은 어휘, 셀 값은 해당 단어의 등장 횟수를 나타낸다.

#### 딥러닝 언어 모델

```text
문장 → 토큰화 → 토큰 ID → 임베딩 벡터 → 딥러닝 모델
```

> **ID(Identifier):** 어휘 사전에서 각 토큰을 구분하기 위해 부여한 정수 번호이다.

> **Embedding:** 토큰 ID처럼 이산적인 값을 의미 관계를 학습할 수 있는 저차원 밀집 벡터로 변환하는 방법이다.

- 토큰 ID는 어휘 사전에서 토큰의 위치를 나타내는 정수 인덱스이며, 의미를 담은 밀집 벡터와 구분된다.
- 임베딩 계층은 토큰 ID를 밀집 벡터로 변환한다.
- 임베딩 결과는 일반적으로 `(배치 크기, 문장 길이, 임베딩 차원)` 형태가 된다.

### 07-02. 벡터화 기법 - 무엇을 하나의 벡터로 표현하는가?

#### 문서 또는 문장 단위

- 문서나 문장 전체를 하나의 벡터로 표현한다.
- 문서 분류, 검색, 문장 유사도 계산 등에 사용한다.
- 대표적인 예는 BoW, TF-IDF, 문장 임베딩이다.

#### 토큰 단위

- 단어 또는 서브워드마다 하나의 벡터를 만든다.
- 단어의 의미 관계를 학습하거나 문장 내부의 문맥을 처리할 때 사용한다.
- 대표적인 예는 One-Hot Encoding, Word2Vec, FastText, BERT의 토큰 임베딩이다.

> **One-Hot Encoding:** 어휘의 각 토큰에 고유한 위치를 배정하고 해당 위치만 1, 나머지는 0으로 표현하는 방법이다.

> **Word2Vec(Word to Vector):** 주변 단어 예측을 통해 단어마다 하나의 고정된 밀집 벡터를 학습하는 기법이다.

> **FastText:** 단어를 문자 단위 조각으로 나누어 학습하므로 미등록 단어나 철자 변형에도 대응할 수 있는 임베딩 기법이다.

> **BERT(Bidirectional Encoder Representations from Transformers):** Transformer 인코더로 양방향 문맥을 학습해 문맥 의존적인 표현을 만드는 사전 학습 언어 모델이다.

아래 그림은 각 토큰을 어휘 크기와 같은 길이의 벡터로 바꾸고, 해당 토큰의 위치만 1로 표시하는 One-Hot Encoding을 보여 준다. 서로 다른 토큰의 벡터가 모두 직교하므로 의미 유사도를 직접 표현하지 못한다.

표현 단위와 벡터 생성 방식은 별개의 기준이다. 예를 들어 BERT는 각 토큰의 벡터를 만들 수 있고, 토큰 벡터를 모아 문장 전체를 나타내는 벡터도 만들 수 있다.

### 07-03. 벡터 값 생성 방법

| 구분 | 대표 기법 | 벡터 특징 | 장점 | 한계 |
|---|---|---|---|---|
| 희소 표현 | One-Hot, BoW, DTM, TF-IDF | 대부분의 값이 0인 고차원 벡터 | 단순하고 해석하기 쉬움 | 단어 순서와 의미 관계를 충분히 반영하지 못함 |
| 정적 임베딩 | Word2Vec, GloVe, FastText | 학습된 저차원 밀집 벡터 | 단어 사이의 의미 관계를 표현함 | 같은 단어는 문맥이 달라도 같은 벡터를 사용함 |
| 문맥 임베딩 | BERT 계열 | 문장 문맥에 따라 달라지는 밀집 벡터 | 다의어와 문맥을 반영함 | 계산량과 메모리 사용량이 큼 |

> **DTM(Document-Term Matrix):** 행에는 문서, 열에는 단어를 두고 각 셀에 단어의 빈도나 가중치를 저장한 행렬이다.

> **GloVe(Global Vectors for Word Representation):** 코퍼스 전체의 단어 동시 등장 통계를 이용해 단어의 정적 임베딩을 학습하는 기법이다.

- **Word2Vec**은 주변 단어를 예측하는 과정에서 단어 벡터를 학습한다.
- **GloVe**는 코퍼스 전체의 단어 동시 등장 통계를 활용한다.
- **FastText**는 단어 내부의 문자 n-gram을 함께 학습해 미등록 단어와 철자 변형에 대응한다.


- **BERT**는 문장의 양쪽 문맥을 함께 사용해 같은 단어도 문장에 따라 다른 벡터로 표현한다.

아래 그림은 밀집 임베딩 공간에서 단어 사이의 관계가 방향과 거리로 나타날 수 있음을 보여 준다. `man → woman`과 `king → queen`의 방향이 비슷하고, 동사의 시제 변화도 유사한 방향으로 표현되는지를 확인한다.

### 07-04. N-gram은 어디에 속하는가?

N-gram은 독립적인 벡터 형식이라기보다 연속된 토큰이나 문자를 묶어 하나의 특성으로 사용하는 방법이다.

- `나는 자연어를 배운다`의 단어 2-gram은 `나는 자연어를` / `자연어를 배운다`처럼 두 단어씩 묶는다.
- CountVectorizer나 TfidfVectorizer에서 n-gram을 사용하면 일부 단어 순서를 특성에 포함할 수 있다.

> **CountVectorizer:** 여러 문서를 토큰별 등장 횟수로 구성된 문서-단어 행렬로 변환하는 Scikit-learn 도구이다.

> **TfidfVectorizer:** 여러 문서를 TF-IDF 가중치로 구성된 문서-단어 행렬로 변환하는 Scikit-learn 도구이다.

- n이 커질수록 더 긴 문맥을 표현하지만 특성 수와 희소성이 함께 증가한다.

### 07-05. 같은 단어가 문맥에 따라 달라지는 예

두 문장 `배를 먹었다`와 `배를 탔다`에는 같은 형태의 `배`가 등장하지만 의미는 서로 다르다.

| 기법 | 만들어지는 표현 | `배`를 다루는 방식 |
|---|---|---|
| TF-IDF | 문장마다 하나의 희소 벡터 | 문장별 중요도는 달라질 수 있지만 `배`의 두 의미를 직접 구분하지 못함 |
| Word2Vec | 단어마다 하나의 정적 벡터 | 두 문장의 `배`에 같은 벡터를 사용함 |
| BERT | 문맥이 반영된 토큰 벡터 | 주변 단어를 이용해 음식과 탈것의 의미를 서로 다르게 표현함 |

### 07-06. 어떤 방법을 선택하는가?

- 빠르고 해석하기 쉬운 텍스트 분류 기준 모델이 필요하면 **TF-IDF와 선형 모델**을 먼저 사용한다.
- 단어 유사도나 가벼운 단어 임베딩이 필요하면 **Word2Vec·GloVe·FastText**를 고려한다.
- 문맥, 다의어, 질의응답, 개체명 인식처럼 문장 의미가 중요하면 **BERT 계열 모델**을 고려한다.

### 핵심 정리

- 문서 단위와 토큰 단위는 **무엇을 표현하는가**에 대한 구분이다.
- 희소 표현, 정적 임베딩, 문맥 임베딩은 **어떻게 숫자를 만드는가**에 대한 구분이다.
- 모든 문제에 가장 복잡한 벡터화가 필요한 것은 아니며 데이터 크기, 과제와 계산 비용에 맞춰 선택한다.
